```python
import numpy as np
import matplotlib.pyplot as plt

# 您提供的 Sigmoid 函数
def sigmoid(x):
  # Our activation function: f(x) = 1 / (1 + e^(-x))
  return 1 / (1 + np.exp(-x))
```

```python
# 1. 生成 x 值
# 我们创建一组从 -10 到 10 的值，以便观察 Sigmoid 函数的完整形状
x_values = np.linspace(-10, 10, 100) # 生成 100 个点

# 2. 计算对应的 y 值
y_values = sigmoid(x_values)

# 3. 使用 Matplotlib 绘制图像
plt.figure(figsize=(8, 6)) # 可以调整图像大小
plt.plot(x_values, y_values, label='sigmoid(x)')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True) # 添加网格线
plt.axhline(0, color='black', linewidth=0.5) # 添加 x 轴
plt.axvline(0, color='black', linewidth=0.5) # 添加 y 轴
plt.axhline(0.5, color='red', linestyle='--', linewidth=0.8, label='y=0.5') # 突出 y=0.5
plt.axhline(1, color='gray', linestyle='--', linewidth=0.8, label='y=1.0') # 突出 y=1.0
plt.legend() # 显示图例
plt.show()
```
